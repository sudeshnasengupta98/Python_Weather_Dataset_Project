# 16) Line chart showing the temperature over time:

```python
# Convert 'Date/Time' column to datetime format
data['Date/Time'] = pd.to_datetime(data['Date/Time'])

# Plotting the temperature over time
plt.figure(figsize=(10, 6))
plt.plot(data['Date/Time'], data['Temp_C'], color='blue')
plt.title('Temperature Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

# 17) Line chart showing the variation of humidity over time:
 

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date/Time'], data['Rel Hum_%'], color='green')
plt.title('Humidity Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Relative Humidity (%)')
plt.grid(True)
plt.show()
```

# 18) Bar chart showing the distribution of weather conditions:


```python
plt.figure(figsize=(10, 6))
data['Weather'].value_counts().plot(kind='bar', color='orange')
plt.title('Distribution of Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

# 19) Box plot showing the distribution of temperature for each weather condition:


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Weather', y='Temp_C', data=data, palette='coolwarm')
plt.title('Temperature Distribution by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()

```

# 20) Bar Plot of Pressures

```python
plt.figure(figsize=(10, 6))
data.groupby('Weather')['Press_kPa'].mean().plot(kind='bar', color='lightcoral')
plt.title('Average Atmospheric Pressure by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Pressure (kPa)')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

# 21) Box Plot of Visibility

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Weather', y='Visibility_km', data=data, palette='Set2')
plt.title('Visibility by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Visibility (km)')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

# 22) Calculate the value counts for each weather condition

```python
weather_counts = data['Weather'].value_counts()

# Calculate the percentages
total = weather_counts.sum()
percentages = [(count / total) * 100 for count in weather_counts]

# Create a pie chart for weather conditions
plt.figure(figsize=(12, 8))

# Plot the pie chart with percentage labels outside
pie_chart = plt.pie(weather_counts, colors=sns.color_palette('pastel'), startangle=90, pctdistance=0.85)

# Create a separate legend for the weather conditions on the side
legend_labels = [f'{weather} ({percent:.1f}%)' for weather, percent in zip(weather_counts.index, percentages)]
plt.legend(legend_labels, loc='center left', bbox_to_anchor=(1, 0.5), title='Weather Conditions')

# Set the title
plt.title('Distribution of Weather Conditions')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show the plot
plt.show()
```
